## Digital IIR Filter Design (Chebyshev Type I)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import cheby1, bilinear, freqz

# Filter specs
analog_order = 4
rp = 1  # Passband ripple in dB
analog_cutoff = 2 * np.pi * 200  # Analog cutoff frequency in rad/s

# Analog filter design
b, a = cheby1(analog_order, rp, analog_cutoff, btype='low', analog=True)

# Digital conversion
Fs = 1000  # Sampling rate in Hz
b_digital, a_digital = bilinear(b, a, fs=Fs)

# Frequency response
w, h = freqz(b_digital, a_digital, worN=8000, fs=Fs)

# Plotting
plt.figure(figsize=(8, 4))
plt.plot(w, 20 * np.log10(abs(h)), 'b')
plt.title('Digital IIR Filter Frequency Response (Chebyshev Type I)')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude (dB)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import cheby2, bilinear, freqz

# Filter specs
analog_order = 4
rs = 40  # Stopband attenuation in dB
analog_cutoff = 2 * np.pi * 200  # Analog cutoff frequency in rad/s

# Analog filter design
b, a = cheby2(analog_order, rs, analog_cutoff, btype='low', analog=True)

# Digital conversion
Fs = 1000  # Sampling rate in Hz
b_digital, a_digital = bilinear(b, a, fs=Fs)

# Frequency response
w, h = freqz(b_digital, a_digital, worN=8000, fs=Fs)

# Plotting
plt.figure(figsize=(8, 4))
plt.plot(w, 20 * np.log10(abs(h)), 'b')
plt.title('Digital IIR Filter Frequency Response (Chebyshev Type II)')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude (dB)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
