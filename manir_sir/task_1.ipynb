## Task 1

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import sawtooth, square

# Time vectors
t = np.arange(0, 10 * np.pi, 0.02)
X = np.arange(0, 10, 0.01)

plt.figure(figsize=(15, 10))

# Sine wave
plt.subplot(3, 3, 1)
plt.plot(t, np.sin(t))
plt.title('Sine Wave')
plt.xlabel('time (s)')
plt.ylabel('Amplitude')
plt.grid(True)

# Cosine wave
plt.subplot(3, 3, 2)
plt.plot(t, np.cos(t))
plt.title('Cosine Wave')
plt.xlabel('time (s)')
plt.ylabel('Amplitude')
plt.grid(True)

# Sawtooth wave
plt.subplot(3, 3, 3)
plt.plot(t, sawtooth(t))
plt.title('Sawtooth Wave')
plt.xlabel('time (s)')
plt.ylabel('Amplitude')
plt.grid(True)

# Triangular wave
plt.subplot(3, 3, 4)
plt.plot(t, sawtooth(t, 0.5))  # triangular is sawtooth with width=0.5
plt.title('Triangular Wave')
plt.xlabel('time (s)')
plt.ylabel('Amplitude')
plt.grid(True)

# Square wave (60% duty cycle)
plt.subplot(3, 3, 5)
plt.plot(t, square(t, duty=0.6))
plt.title('Square Wave (60% duty cycle)')
plt.xlabel('time (s)')
plt.ylabel('Amplitude')
plt.grid(True)

# Exponential growing
plt.subplot(3, 3, 6)
plt.plot(X, np.exp(X / 2))
plt.title('Exponential Growing Wave')
plt.xlabel('time (s)')
plt.ylabel('Amplitude')
plt.grid(True)

# Exponential decaying
plt.subplot(3, 3, 7)
plt.plot(X, np.exp(-X / 2))
plt.title('Exponential Decaying Wave')
plt.xlabel('time (s)')
plt.ylabel('Amplitude')
plt.grid(True)

plt.tight_layout()
plt.show()



n = np.arange(0, 21)
f = 0.1  # frequency in cycles/sample
omega = 2 * np.pi * f

x_sin = np.sin(omega * n)
x_cos = np.cos(omega * n)

plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)
plt.stem(n, x_sin, 'r')
plt.title('Discrete Sine Signal')
plt.xlabel('n')
plt.ylabel('sin(n)')
plt.grid(True)

plt.subplot(2, 1, 2)
plt.stem(n, x_cos, 'b')
plt.title('Discrete Cosine Signal')
plt.xlabel('n')
plt.ylabel('cos(n)')
plt.grid(True)

plt.tight_layout()
plt.show()
```

## Task 2

```python
n1 = np.arange(-10, 11)

omegas = [0, np.pi / 8, np.pi / 4, np.pi / 2, np.pi]
titles = [f' = {w:.3f}' for w in omegas]

plt.figure(figsize=(10, 12))

for i, omega_0 in enumerate(omegas, start=1):
    y = np.cos(omega_0 * n1)
    plt.subplot(len(omegas), 1, i)
    plt.stem(n1, y)
    plt.title(f'Cosine Signal: {titles[i-1]}')
    plt.xlabel('n')
    plt.ylabel('cos( n)')
    plt.grid(True)

plt.tight_layout()
plt.show()
```
