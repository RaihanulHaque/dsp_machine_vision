```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Case I - Real C and a for different conditions

n = np.arange(-10, 21)
C = 1

plt.figure(figsize=(12, 10))

# Case (i.a): a > 1
a_a1 = 2.0  # Use float instead of int
y_a1 = C * a_a1 ** n
plt.subplot(3, 2, 1)
plt.stem(n, y_a1)
plt.title('Case (i.a): a > 1 (a = 2)')
plt.grid(True)

# Case (i.b): 0 < a < 1
a_a2 = 0.5
y_a2 = C * a_a2 ** n
plt.subplot(3, 2, 2)
plt.stem(n, y_a2)
plt.title('Case (i.b): 0 < a < 1 (a = 0.5)')
plt.grid(True)

# Case (i.c): a = 1
a_a3 = 1.0
y_a3 = C * a_a3 ** n
plt.subplot(3, 2, 3)
plt.stem(n, y_a3)
plt.title('Case (i.c): a = 1')
plt.grid(True)

# Case (i.d): a < -1
a_a4 = -2.0
y_a4 = C * a_a4 ** n
plt.subplot(3, 2, 4)
plt.stem(n, y_a4)
plt.title('Case (i.d): a < -1 (a = -2)')
plt.grid(True)

# Case (i.e): -1 < a < 0
a_a5 = -0.5
y_a5 = C * a_a5 ** n
plt.subplot(3, 2, 5)
plt.stem(n, y_a5)
plt.title('Case (i.e): -1 < a < 0 (a = -0.5)')
plt.grid(True)

# Case (i.f): a = -1
a_a6 = -1.0
y_a6 = C * a_a6 ** n
plt.subplot(3, 2, 6)
plt.stem(n, y_a6)
plt.title('Case (i.f): a = -1')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Case II - Complex C and a, |a| = 1 (on unit circle)

n = np.arange(-20, 20.25, 0.25)
C_mag = 1
C_phi = np.pi / 4
a_mag = 1
a_omega = np.pi / 3

C = C_mag * np.exp(1j * C_phi)
a = a_mag * np.exp(1j * a_omega)
y = C * a ** n

plt.figure(figsize=(10, 4))
plt.stem(n, np.imag(y))
plt.title('Case II: Imaginary part of y(n) = C * a^n, with |a| = 1')
plt.xlabel('n')
plt.ylabel('Imag(y)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Case III - Complex C and a, |a|  1 (growth + oscillation)

n = np.arange(16, 30.1, 0.1)
C_mag = 2
C_phi = np.pi / 3
a_mag = 3
a_sigma = 0.5
a_omega = np.pi / 2

C = C_mag * np.exp(1j * C_phi)
a = a_mag * np.exp(1j * (a_sigma + a_omega))
y = C * a ** n

plt.figure(figsize=(10, 4))
plt.stem(n, np.imag(y))
plt.title('Case III: Imaginary part of y(n) = C * a^n with exponential growth & oscillation')
plt.xlabel('n')
plt.ylabel('Imag(y)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
