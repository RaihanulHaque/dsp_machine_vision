```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Time index (discrete)
n = np.arange(-10, 11)
x1 = np.sin(0.2 * np.pi * n)
x2 = np.cos(0.2 * np.pi * n)
```

## Addition of Discrete Signal

```python
y = x1 + x2

plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.stem(n, x1)
plt.title("Signal x1[n] = sin")

plt.subplot(1, 3, 2)
plt.stem(n, x2)
plt.title("Signal x2[n] = cos")

plt.subplot(1, 3, 3)
plt.stem(n, y)
plt.title("x1[n] + x2[n]")

plt.tight_layout()
plt.show()
```

## Substract two signal

```python
y = x1 - x2

plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.stem(n, x1)
plt.title("Signal x1[n] = sin")

plt.subplot(1, 3, 2)
plt.stem(n, x2)
plt.title("Signal x2[n] = cos")

plt.subplot(1, 3, 3)
plt.stem(n, y)
plt.title("x1[n] - x2[n]")

plt.tight_layout()
plt.show()
```

## Multiplication

```python
y = x1 * x2

plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.stem(n, x1)
plt.title("Signal x1[n] = sin")

plt.subplot(1, 3, 2)
plt.stem(n, x2)
plt.title("Signal x2[n] = cos")

plt.subplot(1, 3, 3)
plt.stem(n, y)
plt.title("x1[n]  x2[n]")

plt.tight_layout()
plt.show()
```

## Division

```python
x2_safe = np.where(x2 == 0, 1e-10, x2)
y = x1 / x2_safe

plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.stem(n, x1)
plt.title("Signal x1[n] = sin")

plt.subplot(1, 3, 2)
plt.stem(n, x2)
plt.title("Signal x2[n] = cos")

plt.subplot(1, 3, 3)
plt.stem(n, y)
plt.title("x1[n]  x2[n]")

plt.tight_layout()
plt.show()
```

## Fold Signal

```python
x1_folded = x1[::-1]
n_folded = -n

plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.stem(n, x1)
plt.title("Original Signal x1[n]")

plt.subplot(1, 3, 2)
plt.stem(n_folded, x1_folded)
plt.title("Folded Signal x1[-n]")

plt.subplot(1, 3, 3)
plt.stem(n, x2)
plt.title("x2[n] for Reference")

plt.tight_layout()
plt.show()
```

## Time Shift by 4

```python
shift = 4
x1_shifted = np.roll(x1, shift)

plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.stem(n, x1)
plt.title("Original Signal x1[n]")

plt.subplot(1, 3, 2)
plt.stem(n, x1_shifted)
plt.title(f"Shifted Signal x1[n-{shift}]")

plt.subplot(1, 3, 3)
plt.stem(n, x2)
plt.title("x2[n] for Reference")

plt.tight_layout()
plt.show()
```

# The last One

```python
n = np.arange(-10, 11)
x = np.array([1, 2, 1, 0, -1, -2, -1, 0, 1, 2, 1] + [0]*10)
y = np.array([0]*10 + [2, 3, 2, 0, -1, -1, 0, 1, 2, 1, 0])
# x = np.sin(0.1 * np.pi * n)
# y = np.cos(0.1 * np.pi * n)  
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.stem(n, x)
plt.title("Signal x[n]")
plt.xlabel("n")
plt.ylabel("Amplitude")
plt.subplot(1, 2, 2)
plt.stem(n, y)
plt.title("Signal y[n]")
plt.xlabel("n")
plt.ylabel("Amplitude")
plt.tight_layout()
plt.show()
```

## Auto Correlation

```python
auto_corr = np.correlate(x, x, mode='full')
lags = np.arange(-len(x)+1, len(x))
print("Autocorrelation Rxx[m]:", auto_corr)
print("Lags:", lags)

plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.stem(n, x)
plt.title("Original Signal x[n]")

plt.subplot(1, 3, 2)
plt.stem(lags, auto_corr)
plt.title("Autocorrelation Rxx[m]")

plt.subplot(1, 3, 3)
plt.plot(lags, auto_corr)
plt.title("Autocorrelation (Line Plot)")
plt.grid(True)

plt.tight_layout()
plt.show()
```

## Cross Correlation

```python
cross_corr = np.correlate(x, y, mode='full')
lags = np.arange(-len(x)+1, len(x))

plt.figure(figsize=(12, 4))
plt.subplot(1, 4, 1)
plt.stem(n, x)
plt.title("Signal x[n]")

plt.subplot(1, 4, 2)
plt.stem(n, y)
plt.title("Signal y[n]")

plt.subplot(1, 4, 3)
plt.stem(lags, cross_corr)
plt.title("Cross-Correlation Rxy[m]")

plt.subplot(1, 4, 4)
plt.plot(lags, cross_corr)
plt.title("Cross Correlation with smoother")

plt.tight_layout()
plt.show()
```
