```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(-2, 2, 1000)

# Define continuous signals
signals = {
    "1. Unit Step": np.where(t >= 0, 1, 0),
    "2. Unit Ramp": np.where(t >= 0, t, 0),
    "3. Exponential (a=2)": np.exp(2 * t),
    "4. Exponential Decay (a=-2)": np.exp(-2 * t),
    "5. Sine Wave": np.sin(2 * np.pi * t),
    "6. Cosine Wave": np.cos(2 * np.pi * t),
    "7. Square Wave": np.sign(np.sin(2 * np.pi * t)),
    "8. Sawtooth Wave": 2 * (t - np.floor(t + 0.5)),
    "9. Triangular Wave": 2 * np.abs(2 * (t / 2 - np.floor(t / 2 + 0.5))) - 1,
    "10. Gaussian Pulse": np.exp(-t**2 / 0.1)
}

plt.figure(figsize=(20, 8))
for i, (name, sig) in enumerate(signals.items(), 1):
    plt.subplot(2, 5, i)
    plt.plot(t, sig)
    plt.title(name, fontsize=10)
    plt.grid(True)
plt.tight_layout()
plt.suptitle("Continuous-Time Signals (2x5 Layout)", fontsize=16, y=1.05)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

n = np.arange(-20, 21)  # Discrete time axis

# Discrete counterparts of the continuous signals
discrete_signals = {
    "1. Unit Step": np.where(n >= 0, 1, 0),
    "2. Unit Ramp": np.where(n >= 0, n, 0),
    "3. Exponential (a=2)": np.exp(2 * n),
    "4. Exponential Decay (a=-2)": np.exp(-2 * n),
    "5. Sine Wave": np.sin(2 * np.pi * n / 20),
    "6. Cosine Wave": np.cos(2 * np.pi * n / 20),
    "7. Square Wave": np.sign(np.sin(2 * np.pi * n / 20)),
    "8. Sawtooth Wave": 2 * (n/20 - np.floor(n/20 + 0.5)),
    "9. Triangular Wave": 2 * np.abs(2 * (n/20 - np.floor(n/20 + 0.5))) - 1,
    "10. Gaussian Pulse": np.exp(-n**2 / 10)
}

plt.figure(figsize=(20, 8))
for i, (name, sig) in enumerate(discrete_signals.items(), 1):
    plt.subplot(2, 5, i)
    plt.stem(n, sig)
    plt.title(name, fontsize=10)
    plt.grid(True)
plt.tight_layout()
plt.suptitle("Discrete-Time Signals (2x5 Layout)", fontsize=16, y=1.05)
plt.show()
```
