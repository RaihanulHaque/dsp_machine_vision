```python
import matplotlib.pyplot as plt
import numpy as np
```

## Simple Sine Wave

```python
def plot_sine(freq=1, duration=2, sample_rate=1000):
    t = np.linspace(0, duration, int(sample_rate * duration), endpoint=False)
    y = np.sin(2 * np.pi * freq * t)
    plt.figure(figsize=(8, 3))
    plt.plot(t, y, label='Continuous Sine Wave')
    plt.title('Continuous Sine Wave')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()

plot_sine()
```

## Simple Cos Wave

```python
def plot_cos(freq=1, duration=2, sample_rate=1000):
    t = np.linspace(0, duration, int(sample_rate * duration), endpoint=False)
    y = np.cos(2 * np.pi * freq * t)
    plt.figure(figsize=(8, 3))
    plt.plot(t, y, label='Continuous Cosine Wave', color='orange')
    plt.title('Continuous Cosine Wave')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()

plot_cos()
```

## Discrete Sine Wave

```python
def plot_discrete_sine(freq=1, duration=2, sample_rate=10):
    t = np.linspace(0, duration, int(sample_rate * duration), endpoint=False)
    y = np.sin(2 * np.pi * freq * t)
    plt.figure(figsize=(8, 3))
    plt.stem(t, y, basefmt=" ")
    plt.title('Discrete Sine Wave (Sampled)')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

plot_discrete_sine()
```

## Discrete Cos Wave

```python
def plot_discrete_cos(freq=1, duration=2, sample_rate=10):
    t = np.linspace(0, duration, int(sample_rate * duration), endpoint=False)
    y = np.cos(2 * np.pi * freq * t)
    plt.figure(figsize=(8, 3))
    plt.stem(t, y, basefmt=" ", linefmt='orange', markerfmt='ro')
    plt.title('Discrete Cosine Wave (Sampled)')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

plot_discrete_cos()
```

## Multiplication of Sine and Discrete Sine

```python
def plot_multiplication_sine(freq=1, duration=2, sample_rate=1000, sample_rate_discrete=10):
    t_cont = np.linspace(0, duration, int(sample_rate * duration), endpoint=False)
    sine_cont = np.sin(2 * np.pi * freq * t_cont)

    t_disc = np.linspace(0, duration, int(sample_rate * duration), endpoint=False)
    sine_disc = np.sin(2 * np.pi * freq * t_disc)

    mask = np.zeros_like(t_cont, dtype=bool)
    sample_indices = np.arange(0, len(t_cont), int(sample_rate / sample_rate_discrete))
    mask[sample_indices] = True
    sampled_signal = np.zeros_like(t_cont)
    sampled_signal[mask] = sine_disc[mask]

    multiplied = sine_cont * sampled_signal

    plt.figure(figsize=(8, 3))
    plt.plot(t_cont, multiplied, label='Multiplication Output')
    plt.title('Multiplication of Continuous and Sampled Sine Wave')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()

plot_multiplication_sine()
```

## Multiplication of Cos and Discrete Cos

```python
def plot_multiplication_cos(freq=1, duration=2, sample_rate=1000, sample_rate_discrete=10):
    t_cont = np.linspace(0, duration, int(sample_rate * duration), endpoint=False)
    cos_cont = np.cos(2 * np.pi * freq * t_cont)

    t_disc = np.linspace(0, duration, int(sample_rate * duration), endpoint=False)
    cos_disc = np.cos(2 * np.pi * freq * t_disc)

    mask = np.zeros_like(t_cont, dtype=bool)
    sample_indices = np.arange(0, len(t_cont), int(sample_rate / sample_rate_discrete))
    mask[sample_indices] = True
    sampled_signal = np.zeros_like(t_cont)
    sampled_signal[mask] = cos_disc[mask]

    multiplied = cos_cont * sampled_signal

    plt.figure(figsize=(8, 3))
    plt.plot(t_cont, multiplied, label='Multiplication Output', color='orange')
    plt.title('Multiplication of Continuous and Sampled Cosine Wave')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()

plot_multiplication_cos()
```
